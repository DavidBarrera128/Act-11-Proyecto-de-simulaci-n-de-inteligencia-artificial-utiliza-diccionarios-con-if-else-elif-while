# Clase : Electricista: Certificación de seguridad, zona de cobertura, voltajes manejados, equipo de protección, costo por hora y proyectos activos.


```python
# ====================================================================
# PROGRAMA: SISTEMA IA - GESTIÓN Y RECOMENDACIÓN DE ELECTRICISTAS
# Nivel: Preparatoria (Uso de Funciones, Listas y Diccionarios)
# ====================================================================

# Al inicio, nuestra lista de diccionarios de "electricistas" empieza vacía
electricistas = {}

def registrar_electricista():
    """Función para registrar un nuevo electricista con sus atributos (Key-Value)"""
    print("\n⚡ {REGISTRO DE NUEVO ELECTRICISTA} ⚡")

    # Entrada de datos interactiva (Keys y Values)
    nombre = input("Nombre del electricista: ")
    certificacion = input("¿Cuenta con Certificación de seguridad? (Si/No): ")
    zona = input("Zona de cobertura (ej. Norte, Centro, Sur): ")
    voltaje = input("Voltajes manejados (ej. 110V, 220V, 440V): ")
    equipo = input("Equipo de protección esencial (ej. Guantes, Casco): ")
    costo = input("Costo por hora (Solo números enteros, ej. 250): ")
    proyectos = input("Proyectos activos actualmente (Solo números, ej. 2): ")

    # Creamos el diccionario para ESTE electricista específico
    nuevo_electricista = {
        "Nombre": nombre,
        "Certificación de seguridad": certificacion,
        "Zona de cobertura": zona,
        "Voltajes manejados": voltaje,
        "Equipo de protección": equipo,
        "Costo por hora": costo,
        "Proyectos activos": proyectos
    }

    # Agregamos el diccionario a nuestra lista global en plural: electricistas
    electricistas.append(nuevo_electricista)
    print(f"\n✅ ¡{nombre} ha sido registrado exitosamente en el sistema!")


def mostrar_tabla_electricistas():
    """Función para mostrar los datos bien alineados en forma de tabla"""
    if len(electricistas) == 0:
        print("\n⚠️ La base de datos está vacía. ¡Registra a un electricista primero!")
    else:
        print("\n" + "=" * 125)
        # Encabezados de la tabla con distribución limpia usando formato de alineación (<)
        print(f"{'Nombre':<15} | {'Certificación':<15} | {'Zona Cobertura':<15} | {'Voltajes':<12} | {'Equipo Protección':<22} | {'Costo/Hora':<12} | {'Proy. Activos':<10}")
        print("=" * 125)

        # Ciclo FOR para recorrer la lista de diccionarios
        for e in electricistas:
            print(f"{e['Nombre']:<15} | {e['Certificación de seguridad']:<15} | {e['Zona de cobertura']:<15} | {e['Voltajes manejados']:<12} | {e['Equipo de protección']:<22} | ${e['Costo por hora']:<11} | {e['Proyectos activos']:<10}")
        print("=" * 125)


def asistente_ia():
    """Función que simula una IA (Sistema Experto) usando condicionales recurrentes"""
    if len(electricistas) == 0:
        print("\n🤖 {IA}: Mi base de conocimientos está vacía. No puedo calcular recomendaciones.")
    else:
        print("\n🧠 {BIENVENIDO AL ASISTENTE DE INTELIGENCIA ARTIFICIAL} 🧠")
        print("Dime qué necesitas y buscaré la mejor opción de forma inteligente:")
        print("1. Buscar por Alta Tensión (Voltajes como 440V)")
        print("2. Buscar opción Económica (Bajo presupuesto)")
        print("3. Buscar por Disponibilidad Inmediata (Menos proyectos)")

        seleccion = input("Elige qué criterio evaluar (1-3): ")

        # Simulación de toma de decisiones de la IA con IF, ELIF, ELSE
        if seleccion == "1":
            print("\n🤖 {IA} Analizando certificaciones y voltajes pesados...")
            encontrado = False
            for e in electricistas:
                # Si maneja 440V y además está certificado, la IA lo prioriza
                if "440V" in e["Voltajes manejados"] and e["Certificación de seguridad"].lower() == "si":
                    print(f"💡 RECOMENDACIÓN IA: Te sugiero a {e['Nombre']}. Maneja 440V, está certificado y usa {e['Equipo de protección']}.")
                    encontrado = True
            if not encontrado:
                print("🤖 {IA}: No encontré ningún perfil con certificación experta en 440V en este momento.")

        elif seleccion == "2":
            limite = input("¿Cuál es tu presupuesto máximo por hora? $: ")
            print(f"\n🤖 {IA} Filtrando electricistas accesibles...")
            encontrado = False
            for e in electricistas:
                # Comparamos transformando temporalmente a entero (sin usar try)
                if int(e["Costo por hora"]) <= int(limite):
                    print(f"💡 RECOMENDACIÓN IA: {e['Nombre']} es una opción económica. Cobra ${e['Costo por hora']} por hora en la zona {e['Zona de cobertura']}.")
                    encontrado = True
            if not encontrado:
                print("🤖 {IA}: Lo siento, ningún electricista se ajusta a ese presupuesto tan bajo.")

        elif seleccion == "3":
            print("\n🤖 {IA} Buscando personal libre de proyectos...")
            encontrado = False
            for e in electricistas:
                # Si tiene 0 o 1 proyecto activo, está muy disponible
                if int(e["Proyectos activos"]) <= 1:
                    print(f"💡 RECOMENDACIÓN IA: {e['Nombre']} tiene alta disponibilidad (Solo {e['Proyectos activos']} proyecto(s) activo(s)).")
                    encontrado = True
            if not encontrado:
                print("🤖 {IA}: Todos nuestros electricistas están saturados de trabajo ahora mismo.")

        else:
            print("❌ Opción de IA no reconocida. Intenta de nuevo.")


# ====================================================================
# CICLO PRINCIPAL INTERACTIVO (MENU)
# ====================================================================
while True:
    print("\n⚙️  SISTEMA DE GESTIÓN INTERACTIVO: ELECTRICISTAS ⚙️")
    print("1. Registrar un nuevo electricista")
    print("2. Ver listado completo (Tabla)")
    print("3. Consultar al Asistente IA (Recomendaciones)")
    print("4. Salir del programa")

    opcion_menu = input("Por favor, selecciona una opción (1-4): ")

    # Evaluación de la opción seleccionada del menú
    if opcion_menu == "1":
        registrar_electricista()
    elif opcion_menu == "2":
        mostrar_tabla_electricistas()
    elif opcion_menu == "3":
        asistente_ia()
    elif opcion_menu == "4":
        print("\n🤖 {IA}: Gracias por utilizar el sistema simulado. ¡Hasta la próxima clase!")
        break # Rompe el ciclo while True para cerrar el programa
    else:
        print("❌ Opción inválida. Elige un número del 1 al 4.")
print("Programa realizado por David Barrera 1388")
```
